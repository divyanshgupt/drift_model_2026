```python
import numpy as np
from matplotlib import pyplot as plt
import sys, os
sys.path.append("../../src/")


import scipy.stats as stats
from tqdm import tqdm
from network_time_varying import TimeVaryingFeedForward
```

```python
plt.rcParams["axes.spines.top"] = False
plt.rcParams["axes.spines.right"] = False
plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'Arial'
```

```python

hebb_scaling_range = np.arange(0, 1.0, 0.1)

rand_scaling = 1
for hebb_scaling in hebb_scaling_range:
        print(f"Hebb scaling: {hebb_scaling:.1f}, Rand scaling: {rand_scaling:.1f}")
        network_inh = TimeVaryingFeedForward(hebb_scaling=hebb_scaling, rand_scaling=rand_scaling, inh="on", time_varying=True)
        network_inh.run()
        network_baseline = TimeVaryingFeedForward(hebb_scaling=hebb_scaling, rand_scaling=rand_scaling, inh="on", time_varying=False)
        network_baseline.run()

        drift_mag_inh, drift_rate_inh, convergence_inh = network_inh.get_metrics(network_inh.N, network_inh.n_days, network_inh.theta_stim, network_inh.POs)

        drift_mag_baseline, drift_rate_baseline, convergence_baseline = network_baseline.get_metrics(network_baseline.N, network_baseline.n_days, network_baseline.theta_stim, network_baseline.POs)

        eo = 2
        fig, axs = plt.subplots(1, 3, figsize=(8, 2.5), dpi=200)
        axs[0].plot(np.arange(1, network_inh.n_days)[::eo], np.median(drift_mag_inh, axis=1)[:-1][::eo], c='red', ls='-', marker='o', ms=4, label='CNO', clip_on=False)
        axs[0].plot(np.arange(1, network_baseline.n_days)[::eo], np.median(drift_mag_baseline, axis=1)[:-1][::eo], c='black', ls='-', marker='o', ms=4, label='Baseline', clip_on=False)
        axs[0].set_ylim([0, 5]); axs[0].set_yticks([0, 5])
        axs[0].set_xlabel("time since start [days]")
        axs[0].set_title(r"Drift magnitude $ \; [\degree]$")
        axs[0].legend(frameon=False, fontsize=8)

        axs[1].plot(np.mean(drift_rate_inh, axis=1)[:-1], c='red', ls='-', marker='o', ms=4, label='CNO', clip_on=False)
        axs[1].plot(np.mean(drift_rate_baseline, axis=1)[:-1], c='black', ls='-', marker='o', ms=4, label='Baseline', clip_on=False)
        axs[1].set_ylim([0, 5]); axs[1].set_yticks([0, 5])
        axs[1].set_xlabel("time since start [days]")
        axs[1].set_title(r"Drift rate $ \; [\degree/day]$")
        axs[1].set_xlim(0, 30)
        axs[1].legend(frameon=False, fontsize=8)

        axs[2].plot(np.arange(1, network_inh.n_days)[::eo], np.median(convergence_inh, axis=1)[::eo], c='red', ls='-', marker='o', ms=4, label='CNO', clip_on=False)
        axs[2].plot(np.arange(1, network_baseline.n_days)[::eo], np.median(convergence_baseline, axis=1)[::eo], c='black', ls='-', marker='o', ms=4, label='Baseline', clip_on=False)
        axs[2].set_ylim([-1, 5]); axs[2].set_yticks([0, 5])
        axs[2].set_xlabel("time since start [days]")
        axs[2].set_title(r"Convergence $ \; [\degree]$")
        axs[2].set_xlim(0, 30)
        axs[2].legend(frameon=False, fontsize=8)
        fig.suptitle(f"Hebb scaling: {hebb_scaling:.1f}, Rand scaling: {rand_scaling:.1f}")
        fig.tight_layout()
        fig.savefig(f"../../results/complete_exp/hebb_vs_stochastic/hebb_{hebb_scaling:.1f}_rand_{rand_scaling:.1f}.png", dpi=400)
```
